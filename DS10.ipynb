```python
numeric_cols = iris.select_dtypes(include=np.number).columns
for col in numeric_cols:
    q1 = iris[col].quantile(0.25)
    q3 = iris[col].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5*IQR
    upper_bound = q3 + 1.5*IQR
    iris = iris[(iris[col]>=lower_bound)&(iris[col]<=upper_bound)]

iris
```

```python
sns.boxplot(data=iris, palette="Set2")

plt.title("Boxplot of Features in the Iris Dataset", fontsize=14)
plt.show()
```
